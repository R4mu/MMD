```python
import pandas as pd
import numpy as np
import os
```

```python
features_autoencoder =  pd.read_csv('/content/drive/MyDrive/Data/prepared data/features_autoencoder.csv')
features_denoise_comb = pd.read_csv('/content/drive/MyDrive/Data/prepared data/features_denoising_comb.csv')
features_denoise_mask = pd.read_csv('/content/drive/MyDrive/Data/prepared data/features_denoising_mask.csv')
features_denoise_gauss = pd.read_csv('/content/drive/MyDrive/Data/prepared data/features_denoising_gaussian.csv')
data2 = pd.read_csv('/content/drive/MyDrive/Data/prepared data/volunteer_2022_labelled.csv')
resonance_features = pd.read_csv('/content/drive/MyDrive/Data/prepared data/resonance_features.csv')
```

```python
resonance_features = resonance_features.rename(columns={'File Path': 'kept_paths'})
```

```python
print(data2.columns)
```

```python
data2.drop(['Comorbidity_Index','SARC-F_Score'], axis=1, inplace=True)
```

```python
data2["volunteer"] = "Volunteer_" + (data2.reset_index().index + 1).astype(str)
```

```python
data2.drop(['Volunteer number'],axis=1, inplace=True)
```

```python
data2
```

```python
df_joined_autoencode = pd.merge(features_autoencoder, data2, on="volunteer", how="inner")
df_joined_denoise_comb = pd.merge(features_denoise_comb, data2, on="volunteer", how="inner")
df_joined_denoise_mask = pd.merge(features_denoise_mask, data2, on="volunteer", how="inner")
df_joined_denoise_gauss = pd.merge(features_denoise_gauss, data2, on="volunteer", how="inner")
```

```python
df_joined_autoencode = pd.merge(df_joined_autoencode, resonance_features, on="kept_paths", how="inner")
df_joined_denoise_comb = pd.merge(df_joined_denoise_comb, resonance_features, on="kept_paths", how="inner")
df_joined_denoise_mask = pd.merge(df_joined_denoise_mask, resonance_features, on="kept_paths", how="inner")
df_joined_denoise_gauss = pd.merge(df_joined_denoise_gauss, resonance_features, on="kept_paths", how="inner")
```

```python
df_joined_denoise_mask
```

```python
output_dir = '/content/drive/MyDrive/Data/prepared data'
os.makedirs(output_dir, exist_ok=True)
output_path = os.path.join(output_dir, 'autoencoder_dataset.csv')
df_joined_autoencode.to_csv(output_path, index=False)
print(f"Features DataFrame saved to {output_path}")

output_dir = '/content/drive/MyDrive/Data/prepared data'
os.makedirs(output_dir, exist_ok=True)
output_path = os.path.join(output_dir, 'de_autoencoder_comb_dataset.csv')
df_joined_denoise_comb.to_csv(output_path, index=False)
print(f"Features DataFrame saved to {output_path}")

output_dir = '/content/drive/MyDrive/Data/prepared data'
os.makedirs(output_dir, exist_ok=True)
output_path = os.path.join(output_dir, 'de_autoencoder_mask_dataset.csv')
df_joined_denoise_mask.to_csv(output_path, index=False)
print(f"Features DataFrame saved to {output_path}")

output_dir = '/content/drive/MyDrive/Data/prepared data'
os.makedirs(output_dir, exist_ok=True)
output_path = os.path.join(output_dir, 'de_autoencoder_gauss_dataset.csv')
df_joined_denoise_gauss.to_csv(output_path, index=False)
print(f"Features DataFrame saved to {output_path}")
```
